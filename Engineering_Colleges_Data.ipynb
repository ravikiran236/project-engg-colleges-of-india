### For All Colleges

```python
institute_name = []
location = []
sector =[]
review_count = []
fee =[]
exam = []
rating =[]

for page in range(1,233):
#     url = 'https://engineering.careers360.com/colleges/list-of-engineering-colleges-in-india'.format(page+1)
    r = requests.get('https://engineering.careers360.com/colleges/list-of-engineering-colleges-in-india'+'?sort=popularity&page='+str(page))
    soup = BeautifulSoup(r.content, 'html.parser')
    content = soup.findAll('div',class_='content')
    rating_thumb = soup.findAll('div',class_='ratingThumb')
    for container in content:
        institute = container.h4.text
        institute_name.append(institute)
        place = container.ul.li.text
        location.append(place)
        if len(container.find_all('li')) > 1:
            sctr = container.find_all('li')[1].text
            sector.append(sctr)
        else:
            sector.append('NA')
        if len(container.find_all('li')) > 2:
            review = container.find_all('li')[2].text
            review_count.append(review)
        else:
            review_count.append('NA')
        if len(container.findAll('ul',{'class':'rank'})) > 1:
            second = container.findAll('ul',{'class':'rank'})[1]
            fees = second.li
            if fees is not None:
                fee.append(fees.text)
            else:
                fee.append(0)
            test = second.a
            if test is not None:
                exam.append(test.text)
            else:
                exam.append('NA')
        else:
            fee.append(0)
            exam.append('NA')
    for rate in rating_thumb:
        rtng = rate.a
        if rtng is not None:
            rating.append(rtng.text)
        else:
            rating.append('No Rating')
```

```python
#Converting data to pandas
import pandas as pd
data = pd.DataFrame({'institute_name':institute_name,'location':location,'sector':sector,'review_count':review_count,'fee' :fee,'exam' :exam,'rating':rating})
```

```python
#Checking the shape of the data
data.shape
```

### Removing unwanted text from columns

```python
institute = [i.split(',') for i in list(data['institute_name'])]
data['institute_name'] = pd.DataFrame(institute)
```

```python
data['review_count'] = data['review_count'].map(lambda x:x.strip(' Reviews'))
```

```python
data['rating'] = data['rating'].map(lambda x:x.strip('Rated '))
```

```python
lcn = [i.split(',')[0] for i in list(data['location'])]
data['location'] = pd.DataFrame(lcn)
```

```python
data['fee'] = data['fee'].str.strip('Fees:')
```
